```python
# Perform the necessary imports
import matplotlib.pyplot as plt
fig = plt.gcf()
fig.set_size_inches(12, 6)
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline

samples=df.loc[:,df.columns[0:14]].values

print
# Create scaler: scaler
scaler = StandardScaler();

# Create a PCA instance: pca
pca = PCA();

# Create pipeline: pipeline
pipeline = make_pipeline(scaler,pca)

# Fit the pipeline to 'samples'
pipeline.fit(samples)

# Plot the explained variances
features = range(pca.n_components_)
plt.bar(features, pca.explained_variance_)
plt.xlabel('PCA feature')
plt.ylabel('variance')
plt.xticks(features)
plt.savefig('/Users/chong97/Desktop/Machine_Learning_1/fifthweek/k.jpg')
plt.show()
print(type(features))
```

```python
import numpy as np
sum_pca=pca.explained_variance_/sum(pca.explained_variance_)
cum_pca=np.zeros(13)
cum_pca[0]=sum_pca[0]
for i in range(1,13):
    cum_pca[i]=cum_pca[i-1]+sum_pca[i]
features=range(0,13)
plt.bar(features,cum_pca)
plt.xlabel('PCA feature')
plt.ylabel('variance')
#plt.ylim(0.9,1)
plt.hlines(0.90, 0, 12, colors = "y", linestyles = "dashed")
plt.hlines(0.80, 0, 12, colors = "r", linestyles = "dashed")
plt.xticks(features)
plt.savefig('/Users/chong97/Desktop/Machine_Learning_1/fifthweek/l.jpg')
plt.show()
```

```python
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
x=df.loc[:,df.columns[0:13]].values
y=(df.loc[:,df.columns[13:14]].values).flatten()
log=LogisticRegression(solver='sag',multi_class='multinomial')
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size= 0.5,random_state= 42)

pipe_log = make_pipeline(scaler,log)
pipe_log.fit(x_train, y_train)
print("Accuracy:")
print(pipe_log.score(x_test, y_test))
y_pred=pipe_log.predict(x_test)
from sklearn.metrics import confusion_matrix, classification_report
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
print("Classificaiton Report:")
print(classification_report(y_test, y_pred))
```

```python
pca_1=PCA(n_components=3)
pipe_log_pca = make_pipeline(scaler,pca_1,log)
pipe_log_pca.fit(x_train, y_train)
pipe_log_pca.score(x_test, y_test)
y_pred=pipe_log_pca.predict(x_test)
print("Accuracy:")
print(pipe_log_pca.score(x_test, y_test))
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
print("Classificaiton Report:")
print(classification_report(y_test, y_pred))
```

```python
from sklearn.svm import SVC
svc=SVC(gamma='auto')

pipe_svc = make_pipeline(scaler,svc)
pipe_svc.fit(x_train, y_train)
pipe_svc.score(x_test, y_test)
y_pred=pipe_svc.predict(x_test)
print("Accuracy:")
print(pipe_svc.score(x_test, y_test))
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
print("Classificaiton Report:")
print(classification_report(y_test, y_pred))
```

```python
pca_2=PCA(n_components=3)
pipe_svc_pca = make_pipeline(scaler,pca_2,svc)
pipe_svc_pca.fit(x_train, y_train)
pipe_svc_pca.score(x_test, y_test)
y_pred=pipe_svc_pca.predict(x_test)

print("Accuracy:")
print(pipe_svc_pca.score(x_test, y_test))
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
print("Classificaiton Report:")
print(classification_report(y_test, y_pred))
```

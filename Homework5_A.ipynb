```python
# Perform the necessary imports

fig = plt.gcf()
fig.set_size_inches(12, 6)
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline

samples=df.loc[:,df.columns[1:31]].values

print
# Create scaler: scaler
scaler = StandardScaler();

# Create a PCA instance: pca
pca = PCA();

# Create pipeline: pipeline
pipeline = make_pipeline(scaler,pca)

# Fit the pipeline to 'samples'
pipeline.fit(samples)

# Plot the explained variances
features = range(pca.n_components_)
plt.bar(features, pca.explained_variance_)
plt.xlabel('PCA feature')
plt.ylabel('variance')
plt.xticks(features)

plt.savefig('/Users/chong97/Desktop/Machine_Learning_1/fifthweek/i.jpg')
plt.show()
```

```python
sum_pca=pca.explained_variance_/sum(pca.explained_variance_)
cum_pca=np.zeros(30)
cum_pca[0]=sum_pca[0]
for i in range(1,30):
    cum_pca[i]=cum_pca[i-1]+sum_pca[i]

plt.bar(features, cum_pca)
plt.xlabel('PCA feature')
plt.ylabel('variance')
plt.ylim(0.9,1)
plt.hlines(0.99, 0, 30, colors = "y", linestyles = "dashed")
plt.xticks(features)

plt.savefig('/Users/chong97/Desktop/Machine_Learning_1/fifthweek/j.jpg')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import cross_val_score

x = df.values[:,1:30]
y = df.values[:,31]

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.15,random_state=42);

from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
LiR=LinearRegression()

from sklearn.pipeline import make_pipeline
import time 

start = time.time()
pip_l = make_pipeline(scaler,LiR)

pip_l.fit(x_train,y_train)
y_train_pred=pip_l.predict(x_train)
y_pred=pip_l.predict(x_test)

print("Train R^2: {}".format(pip_l.score(x_train, y_train)))
print("Test R^2: {}".format(pip_l.score(x_test, y_test)))
rmse = np.sqrt(mean_squared_error(y_test,y_pred))
print("Root Mean Squared Error: {}".format(rmse))

end = time.time()
print( end-start)

r2tr=np.zeros(4)
r2te=np.zeros(4)
r2tr[0]=pip_l.score(x_train, y_train)
r2te[0]=pip_l.score(x_test, y_test)

Rms=np.zeros(2)
Rms_p=np.zeros(2)

Rms[0]=np.sqrt(mean_squared_error(y_test,y_pred))

t1=np.zeros(2)
t2=np.zeros(2)

t1[0]=end-start
```

```python
from sklearn.decomposition import PCA

pca=PCA(n_components=3)

start = time.time()
pip_lp = make_pipeline(scaler,pca,LiR)

pip_lp.fit(x_train,y_train)
y_train_pred=pip_lp.predict(x_train)
y_pred=pip_lp.predict(x_test)

print("Train R^2: {}".format(pip_lp.score(x_train, y_train)))
print("Test R^2: {}".format(pip_lp.score(x_test, y_test)))
rmse = np.sqrt(mean_squared_error(y_test,y_pred))
print("Root Mean Squared Error: {}".format(rmse))

end = time.time()
print(end-start)

r2tr[1]=pip_lp.score(x_train, y_train)
r2te[1]=pip_lp.score(x_test, y_test)

Rms_p[0]=np.sqrt(mean_squared_error(y_test,y_pred))

t2[0]=end-start
```

```python
from sklearn.svm import SVR
svr=SVR(gamma='auto',kernel='rbf')

start = time.time()
pip_sr = make_pipeline(scaler,svr)

pip_sr.fit(x_train,y_train)
y_train_pred=pip_sr.predict(x_train)
y_pred=pip_sr.predict(x_test)

print("Train R^2: {}".format(pip_sr.score(x_train, y_train)))
print("Test R^2: {}".format(pip_sr.score(x_test, y_test)))
rmse = np.sqrt(mean_squared_error(y_test,y_pred))
print("Root Mean Squared Error: {}".format(rmse))

end = time.time()
print(end-start)

r2tr[2]=pip_sr.score(x_train, y_train)
r2te[2]=pip_sr.score(x_test, y_test)

Rms[1]=np.sqrt(mean_squared_error(y_test,y_pred))

t1[1]=end-start
```

```python
start = time.time()
pip_srp = make_pipeline(scaler,pca,svr)

pip_srp.fit(x_train,y_train)
y_train_pred=pip_srp.predict(x_train)
y_pred=pip_srp.predict(x_test)

print("Train R^2: {}".format(pip_srp.score(x_train, y_train)))
print("Test R^2: {}".format(pip_srp.score(x_test, y_test)))
rmse = np.sqrt(mean_squared_error(y_test,y_pred))
print("Root Mean Squared Error: {}".format(rmse))

end = time.time()
print(end-start)

r2tr[3]=pip_srp.score(x_train, y_train)
r2te[3]=pip_srp.score(x_test, y_test)

Rms_p[1]=np.sqrt(mean_squared_error(y_test,y_pred))

t2[1]=end-start
```

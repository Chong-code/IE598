```python
# Perform the necessary imports

fig = plt.gcf()
fig.set_size_inches(12, 6)
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline

samples=df.loc[:,df.columns[1:31]].values

print
# Create scaler: scaler
scaler = StandardScaler();

# Create a PCA instance: pca
pca = PCA();

# Create pipeline: pipeline
pipeline = make_pipeline(scaler,pca)

# Fit the pipeline to 'samples'
pipeline.fit(samples)

# Plot the explained variances
features = range(pca.n_components_)
plt.bar(features, pca.explained_variance_)
plt.xlabel('PCA feature')
plt.ylabel('variance')
plt.xticks(features)

plt.savefig('/Users/chong97/Desktop/Machine_Learning_1/fifthweek/i.jpg')
plt.show()
```

```python
sum_pca=pca.explained_variance_/sum(pca.explained_variance_)
cum_pca=np.zeros(30)
cum_pca[0]=sum_pca[0]
for i in range(1,30):
    cum_pca[i]=cum_pca[i-1]+sum_pca[i]

plt.bar(features, cum_pca)
plt.xlabel('PCA feature')
plt.ylabel('variance')
plt.ylim(0.9,1)
plt.hlines(0.99, 0, 30, colors = "y", linestyles = "dashed")
plt.xticks(features)

plt.savefig('/Users/chong97/Desktop/Machine_Learning_1/fifthweek/j.jpg')
plt.show()
```

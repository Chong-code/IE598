```python
import numpy as np

df1.info()
df=df1.drop(['MEDV'],axis=1)
df.head()
```

```python
df1.dropna(axis=0, how='any', inplace=True)
df1.info()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import cross_val_score

x = df1.values[:,0:26]
y = df1.values[:,26]

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42);

LiR=LinearRegression(normalize=True)
LiR.fit(x_train,y_train)
y_train_pred=LiR.predict(x_train)
y_pred=LiR.predict(x_test)
coef_lr=LiR.coef_
intercept_lr=LiR.intercept_

print(coef_lr)
print(intercept_lr)
print("Train R^2: {}".format(LiR.score(x_train, y_train)))
print("Test R^2: {}".format(LiR.score(x_test, y_test)))
rmse = np.sqrt(mean_squared_error(y_test,y_pred))
print("Root Mean Squared Error: {}".format(rmse))
plt.plot(coef_lr)
plt.xlabel('Linear Regression Model')
plt.ylabel('Coefficient')
plt.ylim([-13,6])
plt.savefig('/Users/chong97/Desktop/Machine_Learning_1/fourth_week/c1.jpg')
plt.show()
```

```python
fig = plt.gcf()
fig.set_size_inches(14, 6)
plt.tick_params(labelsize=15)
ax1 = plt.subplot(1,2,1)
plt.sca(ax1)
#plt.ylim([-30,30])
plt.xlabel('y_train')
plt.ylabel('Residual Error')
plt.plot(y_train,y_train-y_train_pred,marker='.',linestyle='none')
plt.hlines(0, 0, 50, colors = "r", linestyles = "dashed")
plt.hlines(5, 0, 50, colors = "y", linestyles = "dashed")
plt.hlines(-5, 0, 50, colors = "y", linestyles = "dashed")

ax2 = plt.subplot(1,2,2)
plt.sca(ax2)
plt.ylim([-30,30])
plt.xlabel('y_test')
plt.ylabel('Residual Error')
plt.plot(y_test,y_test-y_pred,marker='.',linestyle='none')
plt.hlines(0, 0, 50, colors = "r", linestyles = "dashed")
plt.hlines(5, 0, 50, colors = "y", linestyles = "dashed")
plt.hlines(-5, 0, 50, colors = "y", linestyles = "dashed")
```

```python
from sklearn.linear_model import Ridge
ridge = Ridge(normalize=True)

ridge.alpha=0.15
ridge.fit(x_train,y_train)
y_train_predr=ridge.predict(x_train)
y_predr=ridge.predict(x_test)
coef_ridger=ridge.coef_
intercept_ridger=ridge.intercept_

print(coef_ridger)
print(intercept_ridger)
print("Train R^2: {}".format(ridge.score(x_train, y_train)))
print("Test R^2: {}".format(ridge.score(x_test, y_test)))
rmser = np.sqrt(mean_squared_error(y_test,y_predr))
print("Root Mean Squared Error: {}".format(rmser))

alpha_space = np.logspace(-4, 0, 50)
ridge_scores = []
ridge_scores_std = []

def display_plot(cv_scores, cv_scores_std):
    fig = plt.figure()
    ax = fig.add_subplot(1,1,1)
    ax.plot(alpha_space, cv_scores)

    std_error = cv_scores_std / np.sqrt(10)

    ax.fill_between(alpha_space, cv_scores + std_error, cv_scores - std_error, alpha=0.2)
    ax.set_ylabel('CV Score +/- Std Error')
    ax.set_xlabel('Alpha')
    ax.axhline(np.max(cv_scores), linestyle='--', color='.5')
    ax.set_xlim([alpha_space[0], alpha_space[-1]])
    ax.set_xscale('log')
    plt.show()
    
for alpha in alpha_space:

    ridge.alpha = alpha
    ridge_cv_scores = cross_val_score(ridge,x,y,cv=10)
    ridge_scores.append(np.mean(ridge_cv_scores))
    ridge_scores_std.append(np.std(ridge_cv_scores))

display_plot(ridge_scores, ridge_scores_std)
plt.plot(coef_ridger)
plt.xlabel('Ridge Regression Model')
plt.ylabel('Coefficient')
plt.ylim([-13,6])
plt.savefig('/Users/chong97/Desktop/Machine_Learning_1/fourth_week/c2.jpg')
plt.show()
```
